### The Momentum Quality score is based on the Frog-In-The-Pan (FIP) hypthosis which posits that the market under-reacts to gradual positive news. The purpose of the feature is to capture high return stocks that have a continuous and gradual rise. Additional details regarding this theory can be found at https://alphaarchitect.com/2015/11/23/frog-in-the-pan-identifying-the-highest-quality-momentum-stocks/. The FIP quant metric accounts for two factors: 1)price return and 2)path quality (measured by the difference between the percentage of negative and positive days during a period) 

### The FIP metric is represented by the following equation:

## ID = +-(stock_return)(%up_days - %down_days)

```python
%matplotlib inline

import requests
import pandas as pd 
import matplotlib.pyplot as plt

START_DATE = '2011-01-03'
END_DATE = '2019-04-03'
```

```python
file_name = "data/moving-avg-momentum.csv";
output_file = "data/spy-momentum.csv"
df = pd.read_csv(file_name)
```

```python
df.set_index(['Symbol', 'Date'], inplace=True)
```

```python
df['Momentum_Quality_Monthly'] = (df['Pct_Change_Monthly'] * 100) * (( df['Rolling_Monthly_Mean_Positive_Days'] - (1 - df['Rolling_Monthly_Mean_Positive_Days'])))
df['Momentum_Quality_Yearly'] = (df['Pct_Change_Yearly'] * 100) * (( df['Rolling_Yearly_Mean_Positive_Days'] - (1 - df['Rolling_Yearly_Mean_Positive_Days'])))
```

### SPY Trailing Monthly Return Feature is intended to help capture systematic trends in the market.

```python
spy = pd.read_csv("data/spy-price-returns.csv")
```

```python
spy.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
#df.groupby(level=0)['SPY_Trailing_Month_Return'] = spy['Pct_Change_Monthly']
spy.set_index('Date', inplace=True)
```

```python
spy.head()
```

```python
spy_trailing_month_return = spy.drop(columns=['Symbol', 'High', 'Low', 'Open', 'Close', 'Volume', 'AdjClose', 'Pct_Change_Daily', 'Pct_Change_Yearly'])
```

```python
global spy_df
spy_df = pd.DataFrame()
spy_trailing_month_return['SPY_Trailing_Month_Return'] = spy_trailing_month_return['Pct_Change_Monthly']
spy_trailing_month_return.drop(columns=['Pct_Change_Monthly'], inplace=True)

```

```python
spy_trailing_month_return.reset_index(inplace=True)
spy_trailing_month_return.drop(columns=['Date'], inplace=True)

for symbol, r in df.groupby(level=0):
    tmp = r
    tmp.reset_index(inplace=True)
    tick = pd.merge(tmp, spy_trailing_month_return, how='left', left_index=True, right_index=True)
    spy_df = spy_df.append(tick)
```

```python
spy_df.set_index(['Symbol', 'Date'], inplace=True)
```

```python
spy_df.to_csv(output_file, encoding='utf-8', index=True)
```
